```python
# !pip install itchat
import itchat
from collections import Counter
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType
from pyecharts.charts import Map
from pyecharts.charts import WordCloud
import re
#!pip install jieba
import jieba
```

```python
# 扫码登录微信
itchat.auto_login(hotReload=True, enableCmdQR=2)
```

```python
friends = itchat.get_friends()
friends
```

```python
sexs = map(lambda x:x['Sex'], friends[1:])
sexs = list(sexs)
data = list(Counter(sexs).values())
data
```

```python
bar = Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
bar.add_xaxis(["男", "女", "未知"])
bar.add_yaxis("小帅b微信好友性别分布", data)
bar.render_notebook()
```

```python
province = map(lambda x:x['Province'], friends[1:])
provinces = list(province)
provinces = [i for i in provinces if i != '']
```

```python
province_name = []
nums = []

for i in set(provinces):
    province_name.append(i)
    nums.append(provinces.count(i))
```

```python
map = Map()
map.add("小帅b的微信好友位置分布", 
        [list(z) for z in zip(province_name, nums)], "china")
map.set_global_opts(title_opts=opts.TitleOpts(title="小帅b的微信好友位置分布"))
map.render_notebook()
```

```python
signatures = ''

for friend in friends:
    signature = friend['Signature']
    
    if (signature != None):
        signature = signature.strip().replace('span', '').replace('class', '').replace('emoji', '')
        signature = re.sub(r'(< =".*)', '', signature)
        signature = re.sub(r'(\n.*)', '', signature)
        signatures += (signature)
```

```python
data = ' '.join(jieba.cut_for_search(signatures))
data
```

```python
word_count={}

for word in data.split(' '):
    word_count[word] = word_count.get(word, 0) + 1

data = list(word_count.items())
data
```

```python
wc = WordCloud()
wc.add(series_name="个性签名", data_pair=data, word_size_range=[6, 30])
wc.set_global_opts(
        title_opts=opts.TitleOpts(
            title="微信个性签名", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
wc.render_notebook()
```
