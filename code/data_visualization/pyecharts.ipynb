```python
from pyecharts.charts import Scatter, Bar, Line, Pie, Boxplot, WordCloud
from pyecharts import options as opts
import numpy as np
import pandas as pd
```

# 散点图

```python
x = np.random.rand(100)
y = np.random.rand(100)
(
    Scatter(init_opts=opts.InitOpts(width="800px", height="600px"))
    .add_xaxis(xaxis_data=x)
    .add_yaxis(
        series_name="",
        y_axis=y,
        symbol_size=20,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_series_opts()
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            type_="value", splitline_opts=opts.SplitLineOpts(is_show=True)
        ),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        tooltip_opts=opts.TooltipOpts(is_show=False),
    )
    .render_notebook()
)
```

# 折线图

```python
a = np.linspace(0, 1, 10)
b = np.random.randn(20)
(
    Line()
    .add_xaxis(a)
    .add_yaxis('', b)
    .set_global_opts(title_opts=opts.TitleOpts(title="Line"))
    .render_notebook()
)
```

# 直方图

```python
x = ['A', 'B', 'C', 'D']
y1 = np.random.random(10)
y2 = np.random.random(10)
(
    Bar()
    .add_xaxis(x)
    .add_yaxis("A", list(y1))
    .add_yaxis("B", list(y2))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Bar"),
        toolbox_opts=opts.ToolboxOpts(),
        legend_opts=opts.LegendOpts(is_show=False),
    )
    .render_notebook()
)
```

# 饼图

```python
labels = ['java', 'python', 'go', 'c', 'js']
nums = [35, 66, 43, 22, 11]
(
    Pie()
    .add("", [list(z) for z in zip(labels, nums)])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    .render_notebook()
)
```

# 箱线图

```python
v1 = [
    [850, 740, 900, 1070, 930, 850, 950, 980, 980, 880, 1000, 980],
    [960, 940, 960, 940, 880, 800, 850, 880, 900, 840, 830, 790],
]
v2 = [
    [890, 810, 810, 820, 800, 770, 760, 740, 750, 760, 910, 920],
    [890, 840, 780, 810, 760, 810, 790, 810, 820, 850, 870, 870],
]
c = Boxplot()
c.add_xaxis(["expr1", "expr2"])
c.add_yaxis("A", c.prepare_data(v1))
c.add_yaxis("B", c.prepare_data(v2))
c.set_global_opts(title_opts=opts.TitleOpts(title="BoxPlot-基本示例"))
c.render_notebook()
```

# 词云图

```python
data = [
    ("Java", "999"),
    ("Python", "888"),
    ("Go", "777"),
    ("C++", "688"),
    ("C", "588"),
    ("PHP", "516"),
    ("JS", "515"),
    ("Ruby", "483"),
    ("Perl", "462")
]
(
    WordCloud()
    .add(series_name="热点分析", data_pair=data, word_size_range=[6, 66])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="热点分析", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    .render_notebook()
)
```
