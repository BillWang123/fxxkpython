# Seaborn


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

# 散点图

```python
x = np.random.rand(100)
y = np.random.rand(100)
df = pd.DataFrame({'x':x, 'y':y})
sns.relplot(data=df, kind='scatter')
```

# 折线图

```python
sns.relplot(x="x", y="y", kind="line", data=df);
```

# 箱形图

```python
data = np.random.randint(20, size=(10, 4))
df = pd.DataFrame(data, columns=['one','two','three','four'])
sns.catplot(kind="box", data=df);
```

# 条形图

```python
sns.catplot(kind="bar", data=df);
```

```python
sns.barplot(data=df);
```

# 点线图

```python
sns.catplot(kind="point", data=df);
```

# 直方图

```python
a = np.random.rand(666)
s = pd.Series(a)
sns.distplot(s)
```

```python
sns.distplot(s, kde=False)
```

```python
sns.distplot(s, hist=False)
```
